```python
import os
from datasets import load_dataset
from PIL import Image

def export_hf_dataset():
    print("Downloading dataset from Hugging Face...")
    # Load the dataset
    ds = load_dataset("jojogo9/freshness_of_fruits_and_veges_256")
    
    # Create a master folder for these new downloads
    output_base_dir = "downloaded_hf_dataset"
    os.makedirs(output_base_dir, exist_ok=True)
    
    print(f"Dataset loaded! It contains these splits: {list(ds.keys())}")
    
    # We will just pull from the 'train' split for now
    train_data = ds['train']
    
    # Get the label names (e.g., 'fresh_apple', 'rotten_apple')
    # Hugging Face usually stores these in the features dictionary
    label_names = train_data.features['label'].names
    
    saved_count = 0
    
    print("Exporting images to local folders...")
    for idx, item in enumerate(train_data):
        image = item['image']
        label_id = item['label']
        label_name = label_names[label_id]
        
        # Create a subfolder for this specific class (e.g., downloaded_hf_dataset/rotten_apple)
        class_dir = os.path.join(output_base_dir, label_name)
        os.makedirs(class_dir, exist_ok=True)
        
        # Save the image
        # We save it as a .jpg to keep things standard for YOLO
        image_filename = f"hf_{label_name}_{idx}.jpg"
        image_path = os.path.join(class_dir, image_filename)
        
        # Hugging Face images are PIL objects, so we can save them directly
        # Convert to RGB just in case there are weird color profiles/alpha channels
        if image.mode != 'RGB':
            image = image.convert('RGB')
            
        image.save(image_path)
        saved_count += 1
        
        if saved_count % 500 == 0:
            print(f"Saved {saved_count} images...")

    print(f"\nSuccess! Saved {saved_count} total images into '{output_base_dir}'.")

if __name__ == "__main__":
    export_hf_dataset()
```
